# Placing text on the graph

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[5,7,3,8,4]
```

```python
barplot = plt.bar(x,y)
for bar in barplot:
    yval=bar.get_height() #get_height() returns the height of rectange of each bar which is basically the value of y-axis
    print(yval)
    plt.text(bar.get_x() + bar.get_width()/2.0,yval,int(yval),va="bottom",color="green") #plt.text is used to place text on the graph.To get the value of x we can use get_x() and get_width()
plt.title("sample bar graph")
plt.xlabel("Students")
plt.ylabel("Marks")
```

```python
barplot = plt.bar(x,y)
for bar in barplot:
    yval=bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0,yval,int(yval),va="top",color="black") #Here va is vertical alignment.
plt.title("sample bar graph")
plt.xlabel("Students")
plt.ylabel("Marks")
```

```python
x=[1,2,3,4]
y=[5,6,7,8]
a=[9,10,11,12]
b=[13,14,15,16]
plt.bar(x,y,color="red",label="CSS red")
plt.plot(a,b,color="blue",label="CSS blue",linestyle="dashed")
plt.legend(bbox_to_anchor=(1.0, 1.0))
```

# Sorting 

```python
import matplotlib.pyplot as plt
```

```python
x=["A","B","C","D","E","F"]
y=[5,9,2,8,4,6]
y,x=zip(*sorted(zip(y,x))) #sorted order with respect to y-axis
plt.bar(x,y,color="red",label="Sorted order") #sorted order
plt.legend()
plt.title("Sample bar graph")
plt.xlabel("Marks")
plt.ylabel("Students")
```

```python
x=["A","B","C","D","E","F"]
y=[5,9,2,8,4,6]
c=["black","red","green","yellow","blue","orange"]
y,x=zip(*sorted(zip(y,x))) #sorted order
plt.bar(x,y,color=c,label="Sorted order") #sorted order
plt.legend()
plt.title("Sample bar graph")
plt.xlabel("Marks")
plt.ylabel("Students")
```

```python
x=["K","C","A","R","E"]
y=[5,7,9,3,4]
x,y=zip(*sorted(zip(x,y))) #sorted order with respect to x-axis
plt.bar(x,y,color="red",label="Sorted order") #sorted order
plt.legend()
plt.title("Sample bar graph")
plt.xlabel("Marks")
plt.ylabel("Students")
```

# Multi Line plot for two lines 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
product=pd.DataFrame({"year":[2017,2018,2019,2020,2021,2022], "ProdASales" :[2000,3000,4000,3500,6000,5000],"ProdBSales":[3000,4000,3500,3500,5500,6100]})
print(product)
```

```python
ax1=product.plot("year","ProdASales",kind="line",label="Product A Sales",style="r-*")
product.plot("year","ProdBSales",ax=ax1,kind="line",label="Product B Sales",title="Multi line plot for two lines",style="b--o") #ax:axes object
plt.xlabel("Year")
plt.ylabel("Sales")
```

# Multi line plot for three lines

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
product=pd.DataFrame({"year":[2017,2018,2019,2020,2021,2022], "ProdASales" :[2000,3000,4000,3500,6000,5000],"ProdBSales":[3000,4000,3500,3500,5500,6100],"ProdCSales":[3900,2000,3100,5000,4500,4200]})
print(product)
```

```python
plt.plot(product.year,product.ProdASales,"-o",label="Product A Sales")
plt.plot(product.year,product.ProdBSales,"--+",label="Product B Sales")
plt.plot(product.year,product.ProdCSales,"-.*",label="Product C Sales")
plt.xlabel("Year")
plt.ylabel("Sales of different products")
plt.legend()
```

# Subplot

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
finaldf=pd.DataFrame({"Cities":["Delhi","Mumbai","Bangalore","Chennai"],"2017_rainfall":[45,30,15,10],"2018_rainfall":[25,20,25,50]})
print(finaldf)
```

```python
fig=plt.figure()
ax1=fig.add_subplot(121) #121 means 1row, 2columns and 1st plot
finaldf.plot("Cities","2017_rainfall",ax=ax1,kind="bar",title="2017 Rainfall")
ax2=fig.add_subplot(122) #122 means 1row, 2columns and 2nd plot 
finaldf.plot("Cities","2018_rainfall",ax=ax2,kind="line",title="2018 Rainfall")
```

# Multiple Bar Chart

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
company=["infy","google","wipro","tcs","amazon"]
revenue=[145,108,85,122,265]
profit=[26,45,36,38,15]
xpos=np.arange(len(company))
print(xpos) #xpos=len(company)
```

```python
plt.bar(xpos,revenue,width=0.4,label="Revenue")
plt.bar(xpos+0.4,profit,width=0.4,label="Profit")
plt.title("Revenue/Profit bar chart")
plt.xlabel("Company")
plt.xticks(xpos,company)
plt.ylabel("Revenue")
plt.legend(bbox_to_anchor=(1,1))
```
